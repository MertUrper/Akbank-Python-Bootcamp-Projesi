```python
import csv
import os
import datetime

# Pizza ana sınıfı
class Pizza:
    def __init__(self, description, price):
        self._description = description
        self._price = price

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._price

# Pizza alt sınıfları
class ClassicPizza(Pizza):
    def __init__(self):
        super().__init__("Klasik Pizza", 90.0)


class MargheritaPizza(Pizza):
    def __init__(self):
        super().__init__("Margarita Pizza", 100.0)


class TurkishPizza(Pizza):
    def __init__(self):
        super().__init__("Türk Pizza", 110.0)


class SadePizza(Pizza):
    def __init__(self):
        super().__init__("Sade Pizza", 80.0)
```

```python
# Pizza ana sınıfına bağlı Sos ana sınıfı
class Sos(Pizza):
    def __init__(self, component):
        super().__init__(component.get_description(), component.get_cost())
        self.component = component

    def get_cost(self):
        return self.component.get_cost() + super().get_cost()      
        # pizza ve sos ücretinin çekilmesi
        
    def get_description(self):
        return super().get_description() + ' ' + self.component.get_description()
        # pizza ve sos bilgisinin çekilmesi
        
# Sos alt sınıfları
class Zeytin(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Zeytinli"
        self._price = 8.0


class Mantar(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Mantarlı"
        self._price = 14.0


class Peynir(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Peynirli"
        self._price = 12.0


class Pastırma(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Pastırmalı"
        self._price = 18.0


class Sogan(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Soğanlı"
        self._price = 6.0


class Misir(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Mısırlı"
        self._price = 10.0


class TonBaligi(Sos):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Ton Balıklı"
        self._price = 20.0
```

```python
# Menü listesinin oluşturulması
with open("Menu.txt", "w",encoding='utf-8') as menu:
    menu.write("""* Lütfen Pizzanızı Seçiniz:
1: Klasik Pizza ---> 90.0
2: Margarita ------> 110.0
3: Türk Pizza -----> 100.0
4: Sade Pizza -----> 80.0

* Seçmek İstediğiniz Ekstra Malzemeler:
1: Zeytin ------> 8.0
2: Mantar ------> 14.0
3: Peynir ------> 12.0
4: Pastırma ----> 18.0
5: Soğan -------> 6.0
6: Mısır -------> 10.0
7: Ton Balığı --> 20.0

* Şimdiden Bizi Tercih Ettiğiniz İçin Teşekkür Ederiz!\n""")
    
def menu_list():
    with open("Menu.txt","r",encoding='utf-8') as menu:
        menu_items = menu.readlines()
        for item in menu_items:
            print(item.strip())
```

```python
menu_list()
global pizza
global sos

while True:
    # Kullanıcıdan pizza seçimini isteyelim.
    pizza_choice = input("Lütfen Pizza Seçimi Yapınız[1-2-3-4]: ")

    if pizza_choice == "1":
        pizza = ClassicPizza()
        break
    elif pizza_choice == "2":
        pizza = MargheritaPizza()
        break
    elif pizza_choice == "3":
        pizza = TurkishPizza()
        break
    elif pizza_choice == "4":
        pizza = SadePizza()
        break
    else:
        print("Geçersiz bir pizza seçimi yaptınız.")
        continue


# Kullanıcıdan sos seçimini isteyelim.
while True:
    sos_choice = input("Lütfen Bir Sos Seçimi Yapınız[1-2-3-4-5-6-7]: ")

    if sos_choice == "1":
        sos = Zeytin(pizza)
        break
    elif sos_choice == "2":
        sos = Mantar(pizza)
        break
    elif sos_choice == "3":
        sos = Peynir(pizza)
        break
    elif sos_choice == "4":
        sos = Pastırma(pizza)
        break
    elif sos_choice == "5":
        sos = Sogan(pizza)
        break
    elif sos_choice == "6":
        sos = Misir(pizza)
        break
    elif sos_choice == "7":
        sos = TonBaligi(pizza)
        break
    else:
        print("Geçersiz bir sos seçimi yaptınız.")
        continue

# Verilen siparişin pizza/sos bilgisini ve fiyatı yazdırma

print("---------------------------------------------------")
print(f"Sipariş Özeti: {sos.get_description()}")

last_description = sos.get_description()
total_price = sos.get_cost()

print(f"Toplam fiyat: ₺{total_price}")
print("---------------------------------------------------")


# Müşteri Bilgileri
def get_user_info():
    while True:
        name = input("İsim:                           ")
        if any(char.isdigit() for char in name):
            print("İsimde sayı kullanılamaz. Tekrar deneyin.")
            continue
        else:
            break

    while True:
        tc = input("TC Kimlik Numarası(11 Haneli):  ")
        if len(tc) != 11 and tc.isdigit():
            print("TC Kimlik Numarası 11 Haneden Oluşmalıdır.")
            continue
        elif not tc.isdigit():
            print("TC Kimlik Numarası Sadece Sayılardan Oluşmalıdır.")
        else:
            break

    while True:
        card_number = input("Kredi Kartı Numarası(16 Haneli):")
        if not card_number.isdigit() or len(card_number) != 16:
            print("Kredi kartı numarası 16 haneli olmalıdır ve sadece sayılardan oluşmalıdır.")
            continue
        else:
            break

    while True:
        card_password = input("Kredi Kartı Şifresi(4 Haneli):  ")
        if len(card_password) != 4 or not card_password.isdigit():
            print("Kredi kartı şifresi 4 haneli olmalıdır ve sadece sayılardan oluşmalıdır.")
            continue
        else:
            break

    return name, tc, card_number, card_password


name, tc, card_number, card_password = get_user_info()
order_time = datetime.datetime.now()
order_time = datetime.datetime.ctime(order_time)


# Orders_Database.csv dosyasında yer alıcak sütün isimleri
new_order = [name, tc, card_number, card_password, order_time, last_description, total_price]

file_exists = os.path.isfile("Orders_Database.csv")

# Orders_Database.csv dosyasına verilen sipariş ve müşteri bilgilerinin girilmesi
with open("Orders_Database.csv", mode="a", newline="",encoding='utf-8') as file:
    writer = csv.writer(file)
    if file_exists:
        writer.writerow(new_order)
    else:
        header = (["Name", "Id", "Card Number", "Card Password", "Order Time", "Order Description", "Total Price"])
        writer.writerow(header)
        writer.writerow(new_order)
print("Siparişiniz başarıyla oluşturuldu. Afiyet olsun!")
```
